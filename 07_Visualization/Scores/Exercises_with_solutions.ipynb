### Step 1. Import the necessary libraries

```julia
using CSV
using DataFrames
using Dates
using Plots
using Statistics
```

### Step 2. Create the DataFrame that should look like the one below.

```julia
df = DataFrame(
    "first_name" => ["Jason", "Molly", "Tina", "Jake", "Amy"],
    "last_name" => ["Miller", "Jacobson", "Ali", "Milner", "Cooze"],
    "age" => [42, 52, 36, 24, 73],
    "sex" => ["M", "M", "F", "M", "F"],
    "preTestScore" => [4, 24, 31, 2, 3],
    "postTestScore" => [25, 94, 57, 62, 70]
)

first(df, 5)
```

### Step 3. Create a Scatterplot of preTestScore and postTestScore, with the size of each point determined by age
#### Hint: Don't forget to place the labels

```julia
age = df[!, "age"]
min_age = minimum(age)
max_age = maximum(age)
normalize_age = (age .- min_age) / (max_age - min_age) * 20
scatter(
    df[!, "preTestScore"], df[!, "postTestScore"], 
    markersize=normalize_age
)
xlabel!("Pre-test Score")
ylabel!("Post-test Score")
```

### Step 4. Create a Scatterplot of preTestScore and postTestScore.
### This time the size should be 4.5 times the postTestScore and the color determined by sex

```julia
posttest = df[!, "postTestScore"]
min_post = minimum(posttest)
max_post = maximum(posttest)
normalize_posttest = (posttest .- min_post) / (max_post - min_post) * 4.5

sex_colors = Dict("M" => :red, "F" => :blue)
point_colors = [sex_colors[s] for s in df[!, "sex"]]

scatter(
    df[!, "preTestScore"], df[!, "postTestScore"], 
    markersize = normalize_posttest,
    color = point_colors,
    xlabel = "Pre-test Score",
    ylabel = "Post-test Score"

)
```
